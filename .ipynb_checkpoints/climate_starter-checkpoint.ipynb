# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
# create engine to hawaii.sqlite
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
base = automap_base()
# reflect the tables
base.prepare(engine, reflect=True)
```

```python
# View all of the classes that automap found
base.classes.keys()
```

```python
# Save references to each table
measurement = base.classes.measurement
station = base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# Data in measurement table
engine.execute('SELECT * FROM measurement LIMIT 10').fetchall()
```

```python
# Data in station table
engine.execute('SELECT * FROM station LIMIT 10').fetchall()
```

```python
# Use inspect to get column names
inspector = inspect(engine)
columns_m = inspector.get_columns('measurement')
columns_s = inspector.get_columns('station')
```

```python
for x in columns_m:
    print(x['name'],x['type'])
```

```python
for y in columns_s:
    print(y['name'],y['type'])
```

# Exploratory Precipitation Analysis

```python
# Find the most recent date in the data set.
last_date = session.query(func.max(measurement.date)).first()
last_date
```

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results. 
# Starting from the most recent data point in the database.
d_day = dt.date(2017, 8, 23) - dt.timedelta(days=365)
d_day
```

```python
# Calculate the date one year from the last date in data set.
# last_date = 2017-08-23
# Therefore, year from last_date is: 2016-08-23

# Perform a query to retrieve the data and precipitation scores
# Use .filter to filter dates older than a year before last_date
last_year_data = session.query(measurement.date, measurement.prcp).filter(measurement.date >= '2016-08-23').all()
last_year_data
```

```python
for item in last_year_data:
    if item[0] == '2016-08-23':
        print(item[1])
```

```python
for item in last_year_data:
    if item[1] == None:
        print(item[0])
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
last_year_df = pd.DataFrame(last_year_data)
last_year_df = last_year_df.set_index("date")
last_year_df.head()
```

```python
# Sort the dataframe by date
last_year_sorted = last_year_df.sort_values(by="date")
last_year_sorted.head()
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
last_year_sorted.plot(rot=90)
plt.ylabel("Inches")
plt.tight_layout()
plt.savefig("Images/precipitation.png")
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
last_year_sorted.describe()
```

# Exploratory Station Analysis

```python
# Design a query to calculate the total number stations in the dataset

session.query(station.station).count()
```

```python
# Design a query to find the most active stations (i.e. what stations have the most rows?)
session.query(measurement.station, func.count(measurement.station)).group_by(measurement.station).all()
```

```python
# List the stations and the counts in descending order.
session.query(measurement.station, func.count(measurement.station)).\
group_by(measurement.station).order_by(func.count(measurement.station).desc()).all()
```

```python
# Using the most active station id from the previous query, calculate the lowest, highest, and average temperature.
session.query(func.min(measurement.tobs), func.max(measurement.tobs), func.avg(measurement.tobs)).\
filter(measurement.station == 'USC00519281').all()
```

```python
# Using the most active station id
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram
station_data = session.query(measurement.tobs).filter(measurement.date >= '2016-08-23').\
filter(measurement.station == 'USC00519281').all()
station_data
```

```python
station_data_df = pd.DataFrame(station_data, columns = ["Temperature"])
station_data_df = station_data_df.sort_index()
station_data_df.head()
```

```python
station_data_df.plot(kind='hist', bins=12)
plt.xlabel("Temperature (F)")
plt.tight_layout()
plt.savefig("Images/histogram.png")
plt.show()
```
